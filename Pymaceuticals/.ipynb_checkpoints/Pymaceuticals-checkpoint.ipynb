```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
all_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")
# Display the data table for preview
all_data
```

```python
# Checking the number of mice.
all_mice = len(all_data["Mouse ID"].unique())
all_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
all_data["Mouse ID"].value_counts()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

# From the above I can see mouse g989 is the only mouse with too many entries.  I will create a new DF with the mouse ID as
# Index, then locate all data for mouse g989.  I know it has 13 entries so I will call a big enough head to print all data

dupe_data_mice = all_data.set_index('Mouse ID')
m_g989 = dupe_data_mice.loc['g989', :]
m_g989.head(15)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = dupe_data_mice.drop(index = 'g989')
clean_df = clean_df.reset_index()
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
all_mice_clean = len(clean_df["Mouse ID"].unique())
all_mice_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

drug_gr_mean = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
drug_gr_median = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
drug_gr_var = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
drug_gr_stdev = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
drug_gr_sem = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

drug_group_df = pd.DataFrame({"Mean": drug_gr_mean,
                              "Median": drug_gr_median,
                              "Variance": drug_gr_var,
                              "Standard Dev": drug_gr_stdev,
                              "Standard Error of the Mean": drug_gr_sem,})

drug_group_df
    
    

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

agg_sum = clean_df.groupby('Drug Regimen')[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_sum
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_time = pd.DataFrame(clean_df.groupby('Mouse ID')['Timepoint'].max().sort_values())
max_time = max_time.reset_index()
max_time = max_time.rename(columns={'Timepoint': 'Max Timepoint'})

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
clean_df_max = pd.merge(clean_df, max_time, on='Mouse ID')
clean_df_max
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatments_df = clean_df_max.loc[clean_df_max['Drug Regimen'] == treatment]
    
    # add subset 
    subset_df = treatments_df.loc[treatments_df['Timepoint'] == treatments_df['Max Timepoint']]
    tumors = subset_df["Tumor Volume (mm3)"]
    tumor_vol.append(tumors)
    
    quartiles = tumors.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'IQR for {treatment}')
    
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    print(f'Lower Bound for {treatment}: {lower_bound}')
    print(f'Upper Bound for {treatment}: {upper_bound}')   
    
    # Determine outliers using upper and lower bounds
    outliers = tumors.loc[ (treatments_df['Tumor Volume (mm3)'] >= upper_bound) | 
                           (treatments_df['Tumor Volume (mm3)'] <= lower_bound) ].count()
    print(f'Number of {treatment} outliers: {outliers}')    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
green_diamond = dict(markerfacecolor='g', marker='D')
plt.boxplot(tumor_vol, flierprops=green_diamond)
plt.title('Final Tumor Volumes by Drug')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

# Create a DF of just mouse i557, A mouse treated with Capomulin I picked at random, then plots the mouse's tumor by timepoint
mouse_capo = clean_df_max.loc[clean_df_max['Mouse ID'] == 'i557']

plt.plot(mouse_capo['Timepoint'], mouse_capo['Tumor Volume (mm3)'], marker = 'x', color='g')
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Mouse i557's Tumor progress on Capomulin")
plt.grid()
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_df = clean_df_max.loc[clean_df_max['Drug Regimen'] == 'Capomulin']

# Creates a DF of average tumor volume with corresponding mouse IDs, then merges that with the capomulin_df
av_tumor_vol = pd.DataFrame(capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values())
av_tumor_vol = av_tumor_vol.rename(columns={'Tumor Volume (mm3)': 'Average Tumor Volume'})
av_tumor_vol = av_tumor_vol.reset_index()
av_tumor_vol_df = pd.merge(capomulin_df, av_tumor_vol, on='Mouse ID')

# Assigning the columns to variables as their respective axises, then plots them via plt.scatter
x_axis = av_tumor_vol_df['Weight (g)']
y_axis = av_tumor_vol_df['Average Tumor Volume']
plt.scatter(x_axis, y_axis)
plt.xlabel("Weight (g) of Mouse")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Mouse Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = st.pearsonr(x_axis, y_axis)
print(f'The correlation between Weight and the Average Tumor Volume for the Capomulin regimen is: {round(corr[0],2)}')

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regress_values, "r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="red")
plt.show()
```
